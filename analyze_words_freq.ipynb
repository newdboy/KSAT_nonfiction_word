```python
import copy

def dict_merger(dict1, dict2):
    result = copy.deepcopy(dict1)
    for key, value in dict2.items():
        if key in dict1:
            result[key] = dict1[key] + value
        else:
            result[key] = value
    return result

def dict_over_freq_filter(test, over_freq=5):
    test_ = dict()
    for keyw, freq in test.items():
        if freq > over_freq:
            test_[keyw] = freq
    return test_
```

```python
# re딕셔너링 {keyword: freq}

import re
def redictionaring_from_poskeyw(input_dict):
    tmp = dict()
    for keyw, freq in input_dict.items():
        edited_keyw = re.sub(' ', '', restore_word_from_pos(keyw))
        if edited_keyw in tmp:
            # print('이미 있었네!', edited_keyw, 'from: ', keyw)
            tmp[edited_keyw] = tmp[edited_keyw] + freq
        else:
            # print(keyw, edited_keyw)
            tmp[edited_keyw] = freq
    return tmp
```

```python
# 형태소 분석 결과 추가 정의 (add_pre_analyzed_word)
import pickle

from kiwipiepy import Kiwi
kiwi = Kiwi(num_workers=4, model_type='sbg')  #, typos='basic'
kiwi.add_pre_analyzed_word('실세계', [('실','NNG'), ('세계', 'NNG')], 100)
kiwi.add_pre_analyzed_word('격자판', [('격자', 'NNG'), ('판','NNG')], 100)
kiwi.add_pre_analyzed_word('신용도', [('신용', 'NNG'), ('도','NNG')], 100)
kiwi.add_user_word('내인성', 'NNP', 100)
# for x in noun_word_list:
    # kiwi.add_user_word(x, 'NNP', 100)
```

```python
def tokenizer_kiwi(sent, pre_spacing=False):  # [(형태소1, 품사1), (형태소2, 품사2), ...] 형태로 결과를 리턴
    if pre_spacing:
        sent = kiwi.space(sent, reset_whitespace=True)
    result = list(map(lambda x: (x[0], x[1], x[2], x[2]+x[3]), kiwi.tokenize(sent)))
    result_ = list()
    k=0

    # kiwipiepy(kiwipiepy-0.15.2) bug fix
    while k < len(result):
        if result[k][0]=='격자':
            if result[k+1][0]=='판':
                result_.append((result[k][0], result[k][1], result[k][2], result[k][2]+len(result[k][0])))
                result_.append((result[k+1][0], result[k+1][1], result[k][2]+len(result[k][0]), result[k][2]+len(result[k][0])+len(result[k+1][0])))
                k+=2
            else:
                result_.append(result[k])
                k+=1
        elif result[k][0]=='실':
            if result[k+1][0]=='세계':
                result_.append((result[k][0], result[k][1], result[k][2], result[k][2]+len(result[k][0])))
                result_.append((result[k+1][0], result[k+1][1], result[k][2]+len(result[k][0]), result[k][2]+len(result[k][0])+len(result[k+1][0])))
                k+=2
            else:
                result_.append(result[k])
                k+=1
        elif result[k][0]=='내':
            if result[k+1][0]=='인성':
                result_.append((result[k][0], result[k][1], result[k][2], result[k][2]+len(result[k][0])))
                result_.append((result[k+1][0], result[k+1][1], result[k][2]+len(result[k][0]), result[k][2]+len(result[k][0])+len(result[k+1][0])))
                k+=2
            else:
                result_.append(result[k])
                k+=1
        elif result[k][0]=='신용':
            if result[k+1][0]=='도':
                result_.append((result[k][0], result[k][1], result[k][2], result[k][2]+len(result[k][0])))
                result_.append((result[k+1][0], result[k+1][1], result[k][2]+len(result[k][0]), result[k][2]+len(result[k][0])+len(result[k+1][0])))
                k+=2
            else:
                result_.append(result[k])
                k+=1
        else:
            result_.append(result[k])
            k+=1
        # print(k)
        # print(result[k])
    return result_
```

```python
from collections import defaultdict


def nonposword_freq_calculate_non_gap(word_freq_dict, connected_word_list):
    result = defaultdict(int)
    for word, freq in word_freq_dict.items():
        tmp_word = kiwi.join(word)
        if tmp_word in connected_word_list:
            result[tmp_word] += freq
    return result

def concat_dict(dict1, dict2):
    sumdict = defaultdict(int)
    for x, y in dict1.items():
        sumdict[x] += y
    for x, y in dict2.items():
        sumdict[x] += y
    return sumdict

tbs_2023_vocab = concat_dict(textbook_2015curri_vocab, ebs_2023_vocab)
tbs_2022_vocab = concat_dict(textbook_2015curri_vocab, ebs_2022_vocab)
tbs_2021_vocab = concat_dict(textbook_2015curri_vocab, ebs_2021_vocab)
```
